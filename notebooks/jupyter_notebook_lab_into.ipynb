# A more technical description of Jupyter notebook

Notebook combines three component.

- The notebook web application:
An interactive web application for writing and running the code and documentation.


- Kernel:
A process started by web application. It runs the code in a language(Python R, scala etc.) and returns the result to web application.

- Notebook documents: 
 a documents containing representation of all content(code, input, output) in the notebook.
 Internally, notebook documents are [JSON](https://en.wikipedia.org/wiki/JSON).(data  consisting of key–value pairs )

When we type 

***jupyter notebook or jupyter lab*** in the terminal, it starts
1. A notebook server
1. A Jupyter notebook web application, the dashboard, shows the notebooks in current directory
    - In case of  jupyter lab,  dashboard has files browser, code console etc. It gives more tools for richer
    workflow.

# Dictionaries
Dictionaries allows mapping object to other objects i.e they store key value pairs. They are extremly efficient in look up based on key.

**Key**: value used for index

**Value**: stored value

#  creating dictionaries

```python
algoConfig = {} # can use also algoConfig = dict()
algoConfig['threshold'] = 20 # in pixel
algoConfig['neighborhoodSize'] = 4 # 4X4 neighborhood
algoConfig['tolerance'] = 0.002 # convergence tolerance
print(algoConfig)
```

# How to iterate over dictionary items

```python
for name_key, value in algoConfig.items():
    print('name of paramter is {} and value is {}'.format(name_key, value))
```

# defaultdict
setdefault is useful when key doesn't exist. This can be monotonous we have to do every time.

Let's say we wan to count the frequency of number in in a list

```python
# we can do this way
observ = [1, 3, 5, 7, 3, 7, 7]
def countFreq(observations):
    frequencies = {}
    for item in observations:
        freq = frequencies.setdefault(item, 0)
        frequencies[item] = freq + 1
    return frequencies    
print (countFreq(observ))
```

```python
# or use defaultdict
from collections import defaultdict
def countFreqDefaultDictUsage(observations):
    frequencies = defaultdict(int)
    for item in observations:        
        frequencies[item] = frequencies[item] + 1
    return frequencies    
print (countFreqDefaultDictUsage(observ))
```

# writing function but with variable number of arguments

```python
# see this example
def logInfo(level, *args ):
    print (level)
    print (args)
    for arg in args:
        print(arg)
        
logInfo(1, '9 Oct', 'Houston', 75 )
```

As you can guess after reading the code, variable number of arguments are packed into a tuple.

Using *operator we can do the opposite i.e unpack the argument out of a list or tuple

```python
param = [4, 10]

list(range(*param))
```

# Passing variable keyword argument
Similary for passing variable length keyword argument use following syntax

```python
def logInfoKeyWord(level, **kwargs ):
    print (level)
    print (kwargs)
    for key , val in kwargs.items():
        print('key = {}, value = {}'.format(key, val))
        

logInfoKeyWord(1, Date= '7 Jan', City = 'Denver', Temperature=45)
```

Or we can use ** operator on a dictionary to get keyword arguments

```python
params = {'Date': '9 Nov', 'City' : 'Houston', 'Temperature':75} 
logInfoKeyWord(1, **params )
```

Before the variable number of arguments, zero or more normal arguments may occur. if we use both then args must occur before *kwargs
