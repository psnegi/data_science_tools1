# awk basics
awk works on records. If record seprator is newline then each line is a record.

<center>
<b>
BEGIN {initilization command before processing file}
<br><br>
{command for each record(line).}
<br><br>
END{awk command after reading file}</b> </center>

## How awk operates on a each line
- check any pattern matching instruction(regular expressions as for grep),otherwise it operates on all lines
- perform supported math or string manipulation.

Note: awk also supports associative arrays

## Pre-defined and automatic variables in AWK
Use these variable as required to write program
- **RS**: record seperator. awk process input stream a record at a time. default is newline.
- **NR**: The current input record number. If newline is delimitor, then it is current line number.
- **FS/OFS**: awk splits a record using FS into different fields. When priting a records it joins the field using OFS. Usually FS/OFS are same with default value white space. 


- **NF**: The number of fields in the current record.

- awk splits a record into fields.
    + "\$0" stores whole record.
    + "\$1" stores first field
    + "$2" stores second record and so on.

# Arrays in AWK
arrays in AWK are [associative arrays](https://en.wikipedia.org/wiki/Associative_array). 
In short collcection of (key,value) pairs, like you can use string to retreive the value.

Let see this in action. Say we want to print total number of movie for all the genres

```python
!head movies.csv
```

```python
!awk 'BEGIN{FS="," } NR>1{ gsub(/\r/,"",$NF); n=split($NF,a,"|");   for(i=1;i<=n;i++) {  GENRES[a[i]]++} } END{ for(g in GENRES) print  g,GENRES[g] }' movies.csv
```

**Year of movie is embedded in the title can we extract it?**

We can use regex to find year but how to get the find pattern for output?

special symbol **&**  corresponds to pattern found

# Let find year and put [] (or any other decoration you want to do) around it

```python
!head -n 5 movies.csv| sed  's/([0-9]*)/ [year &]/'
```

## other usage format
sed -e '/pattern or regex/ **command**' file_to_porcess

**command can be**
- 's' = search & replace
- 'p' = print
- 'd' = delete
- 'i' = insert
- 'a' = append



**Let's say we want to print  all the line in a code with tag #TODO**

Note: By default sed will print all the line that do not match. Incase of substitution new line will be printed. To supress this use **-n** flag.

Use **p** to print modified lines.

```python
!echo "#this is done\n#TODO need to try linear model\n#TODO transform date"|sed  -n '/^#TODO/ p'
```
